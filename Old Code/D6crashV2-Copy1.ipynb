### SPF Calculation


```python
#Read in the coefficients Excel file (which follows the format described in the documentation)
coeff = pd.read_excel("Data/Spring 2022 Coefficients.xlsx")
coeff.Label

# Obtain dispersion from the Excel file
dispersion = float(coeff.Estimate[coeff.Label=="Dispersion"])
print(coeff.Label)
print(dispersion)

#Convert coefficients to frequency by dividing by years
years = float(coeff.Estimate[coeff.Label=="Years"])

#Obtain a list of all the fields in the data
factor_list = list(coeff.Label)
factor_list
#Initialize the SPF dictionary.  Each district has a value equal to its intercept plus its log(AADT)*AADT_Coefficient, and all future values are added to these values 
stats_dict = {o: {"Predicted_Before_Original":float(coeff.Estimate[coeff.Label=="(Intercept)"])+float(coeff.Estimate[coeff.Label=="log(AADT)"])*np.log(info_dict[o]["AADT_Before"]), "Predicted_After_Original":float(coeff.Estimate[coeff.Label=="(Intercept)"]) + float(coeff.Estimate[coeff.Label=="log(AADT)"])*np.log(info_dict[o]["AADT_After"])} for o in set(CurveIDs)}

#Obtain route and county coefficients
route_coeff = [float(coeff.Estimate[coeff.Label==f]) for f in factor_list if "route_" in f]
counties_coeff = [float(coeff.Estimate[coeff.Label==f]) for f in factor_list if "C_" in f]

stats_dict = {o: {"Predicted_Before_Original":float(coeff.Estimate[coeff.Label=="(Intercept)"]),
"Predicted_After_Original":float(coeff.Estimate[coeff.Label=="(Intercept)"])} for o in set(CurveIDs)}

#Initialize the SPF dictionary.  Each district has a value equal to its log(AADT)*AADT_Coefficient, and all future values are added to this 
stats_dict = {o: {"Predicted_Before_Original":float(coeff.Estimate[coeff.Label=="log(AADT)"])*np.log(info_dict[o]["AADT_Before"]),
                  "Predicted_After_Original":float(coeff.Estimate[coeff.Label=="log(AADT)"])*np.log(info_dict[o]["AADT_After"])} for o in set(CurveIDs)}


#add all features that do not need to be transformed
for o in set(CurveIDs):
    for factor in factor_list:
        try:
            stats_dict[o]["Predicted_Before_Original"] += float(coeff.Estimate[coeff.Label==factor])*info_dict[o][factor]
            stats_dict[o]["Predicted_After_Original"] += float(coeff.Estimate[coeff.Label==factor])*info_dict[o][factor]
        except:
            continue

#add features that do need to be transformed.  Here, they are all log
log_factors = [factor for factor in factor_list if "log" in factor and "AADT" not in factor]

for o in set(CurveIDs):
    for factor in log_factors:
        try:
            stats_dict[o]["Predicted_Before_Original"] += float(coeff.Estimate[coeff.Label==factor])*np.log(info_dict[o][factor])
            stats_dict[o]["Predicted_After_Original"] += float(coeff.Estimate[coeff.Label==factor])*np.log(info_dict[o][factor])
        except:
            continue


#add counties and route coefficients
county_names = [f for f in factor_list if "C_ in f"]
route_names = [f for f in factor_list if "route_" in f]

for id in info_dict:
    county = 0
    route = 0
    truck = 0

    for c in county_names:
        try:
            county+=info_dict[id][c]
        except:
            continue

    for c in route_names:
        try:
            route+=info_dict[id][c]
        except:
            continue

#If we decide to go back to having the possibility of negative predicted crashes, then the first 2 lines in the loop should be uncommented, and the next 2 should be commented
for fid in stats_dict:
    #stats_dict[fid]["Predicted_Before"] = max(stats_dict[fid]["Predicted_Before_Original"], 0)
    #stats_dict[fid]["Predicted_After"] = max(stats_dict[fid]["Predicted_After_Original"], 0)
    stats_dict[fid]["Predicted_Before"] = stats_dict[fid]["Predicted_Before_Original"]
    stats_dict[fid]["Predicted_After"] = stats_dict[fid]["Predicted_After_Original"]


#observed curve before/after.  This uses output, which was computed earlier
for curve in output:
    stats_dict[curve[0]]["Observed_Before"] = curve[1]/(8*date_ratio_dict[1])
    stats_dict[curve[0]]["Observed_After"] = curve[2]/(8*(1-date_ratio_dict[1]))

    
#observed total before/after, ignoring the curve filters.  This uses total_output, which was computed earlier
for curve in total_output:
    if curve[0] in list(stats_dict.keys()):
        stats_dict[curve[0]]["Total_Before"] = curve[1]/(8*date_ratio_dict[1])
        stats_dict[curve[0]]["Total_After"] = curve[2]/(8*(1-date_ratio_dict[1]))

```

### CMF Calculation

```python
#compute CMF's

for fid in stats_dict:
    stats_dict[fid]["w_list"] = [1/(1+dispersion*stats_dict[fid]["Predicted_Before"]), 1/(abs(stats_dict[fid]["Observed_Before"]-stats_dict[fid]["Predicted_Before"])+1)]
#1/(abs(observed-predicted)+1)

#calculate expected before treatment
for fid in stats_dict:
    num = 1
    for w in stats_dict[fid]["w_list"]:
        stats_dict[fid]["Expected_Before" + str(num)] = w*stats_dict[fid]["Predicted_Before"] + (1-w)*stats_dict[fid]["Observed_Before"]
        num+=1

#calculate expected after treatment (if no treatment was applied)
for fid in stats_dict:
    num = 1
    for w in stats_dict[fid]["w_list"]:
        try:
            stats_dict[fid]["Expected_After" + str(num)] = stats_dict[fid]["Expected_Before"+str(num)] * stats_dict[fid]["Predicted_After"] / stats_dict[fid]["Predicted_Before"]
            num+=1
        except:
            stats_dict[fid]["Expected_After" + str(num)] = 0
            num+=1


#calculate variance
for fid in stats_dict:
    num = 1
    for w in stats_dict[fid]["w_list"]:
        try:
            stats_dict[fid]["Variance" + str(num)] = stats_dict[fid]["Expected_After"+ str(num)] * stats_dict[fid]["Predicted_After"] * (1-w) / stats_dict[fid]["Predicted_Before"]
            num+=1
        except:
            stats_dict[fid]["Variance" + str(num)] = 0
            num+=1

#calculate CMF
for fid in stats_dict:
    num = 1
    for w in stats_dict[fid]["w_list"]:
        try:
            cmf_part1 = stats_dict[fid]["Observed_After"]/stats_dict[fid]["Expected_After"+ str(num)]
            cmf_part2 = 1+(stats_dict[fid]["Variance"+ str(num)]/((stats_dict[fid]["Expected_After"+ str(num)])**2))
            stats_dict[fid]["CMF" + str(num)] = cmf_part1/cmf_part2
            num+=1
        except:
            stats_dict[fid]["CMF" + str(num)] = 0
            num+=1

#calculate CRF
for fid in stats_dict:
    num = 1
    for w in stats_dict[fid]["w_list"]:
        stats_dict[fid]["CRF" + str(num)] = 100*(1-stats_dict[fid]["CMF"+ str(num)])
        num+=1
```
